```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather]) #Returns dataframe using ORIGINAL INDEXES
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index=True) #ignore_index=True continues indexing
df
```

```python
df = pd.concat([india_weather, us_weather], keys=["india", "us"]) #keys associates a key with each of the dataframes
df

#Created a index for each subset of the dataframe
```

```python
df.loc["india"] #now we can retrieve a subset of our new dataframe
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df

#Index arguments help us organize by rows
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,11],
}, index=[1,0])
windspeed_df
```

```python
new_df = pd.concat([temperature_df, windspeed_df]) # !WARNING! This way appends the dataframes as ROWS. 
new_df
```

```python
new_df = pd.concat([temperature_df, windspeed_df], axis=1) # axis=1 allows us to seperate the data
new_df
```
