```python
# https://www.youtube.com/watch?v=Wb2Tp35dZ-I&list=PLeo1K3hjS3uuASpe-1LjfG5f14Bnozjwy&index=7
import pandas as pd

df = pd.read_csv('../datasets/weather_by_cities.csv')
df
```

```python
# group based on cities
# First paramter is the column that you want to use for grouping

g = df.groupby('city') # returns a groupby object
# g is a key value object that has created seperate dataframes based on city
# We can access these dataframes by iterating over this object
g
```

```python
for city, dataframe in g:
    print(city)
    print(dataframe)
```

```python
# get a dataframe by key
g.get_group('mumbai')
```

```python
# What is the maximum temperature in EACH of these cities?
g.max()
# We divided up our dataframe into seperate dataframes by city, found the max values in each column by city dataframe
#  and then combined all the data into one single dataframe. Wow! Thats a lot for not a lot of code!

# This process is called "Split Apply Combine"
```
