```python
# https://www.youtube.com/watch?v=XOxABiMhG2U&index=6&list=PLeo1K3hjS3uuASpe-1LjfG5f14Bnozjwy

'''Be careful when running cells, some may not be exemplifying the concept mentioned in the cell comment
    because of the order in which you run the cells. This is especially the case for the regex examples'''

import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/weather_data_replace.csv")
df
```

```python
# two parameters: value to replace and value to replace with
# NOTE: The first parameter can accept a LIST of values to look for and replace with the second parameter
new_df = df.replace([-99999, "0"], np.NaN)
new_df
```

```python
#Replace value by specific column
#If you pass in a DICTIONARY instead of a list for the first parameter
#Key: name of column, Value: value to replace

new_df = df.replace({'temperature': -99999, 'windspeed': -99999, 'event': "0"},np.NAN)
new_df
```

```python
#Map the dataframe with values you want to replace
new_df = df.replace({-99999: np.NAN, "0": "Sunny"})
new_df
```

```python
#ALTERING DATAFRAME FOR REGEX EXAMPLE IN NEXT CELL
new_df = df.replace({'temperature': -99999}, '5 F')
```

```python
new_df = new_df.replace({'windspeed': -99999}, '10 mph')
new_df
```

```python
# The first parameter can also be a regex expression
new_df = new_df.replace('[A-Za-z]', '', regex=True)
new_df
# UH OH! SEVERAL VALUES FROM EVENT COLUMN ARE GONE!
```

```python
# The first parameter can also be a regex expression ON ONLY A COLUMN
new_df = new_df.replace({'temperature': '[A-Za-z]', 'windspeed': '[A-Za-z]'}, '', regex=True)
new_df
```
